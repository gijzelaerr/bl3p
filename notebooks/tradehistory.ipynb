```python
from pybl3p.public import tradehistory
from datetime import datetime
import pandas as pd
%matplotlib inline
```

```python
df = pd.DataFrame(tradehistory(timefactor='d', timevalue=300))
```

```python
df['t'] = pd.to_datetime(df['t'],unit='s')
```

```python
df.set_index('t', inplace=True)
```

```python
df.plot();
```
